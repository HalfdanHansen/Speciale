```python
%matplotlib qt

import pandas as pd

import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib.patches as mpatches
from numpy import random
from matplotlib.font_manager import FontProperties

fig = plt.figure(figsize=[14,7])

fig.suptitle('Papernet results')

col1, col2, col3 = cm.jet(np.array([0.9, 0.1,0.3]))
col4, col5 = cm.Greens(np.array([0.4,0.6]))
fontP = FontProperties()
fontP.set_size('small')


datanormal = pd.read_csv('07_05_Papernet2True_CIFAR10.csv', index_col = False, header = None)
data3d = pd.read_csv('07_05_Papernet2True3D_CIFAR10.csv', index_col = False, header = None)
data3dshared = pd.read_csv('07_05_Papernet2True3DSharedWeights_CIFAR10.csv', index_col = False, header = None)
#===============
#  First subplot
#===============
ax = fig.add_subplot(2, 3, 1)
X = np.arange(0, 50, 1)
ax.plot(X, datanormal[0:50], color=col1)
ax.plot(X, data3d[0:50], color=col2)
ax.plot(X, data3dshared[0:50], color=col3)
ax.plot(X, np.ones(50)*0.1, color=col4)
ax.plot(X, np.ones(50)*0.2, color=col5)
ax.set_title('Training')
#ax.set_xlabel('Epoch')
ax.set_ylabel('Papernet2\nAccuracy')
#ax.text(2, 6, r'Papernet2', fontsize=15)
ax.set_xlim(0, 50)
ax.set_ylim(0, 1)

#===============
# Second subplot
#===============

ax = fig.add_subplot(2, 3, 2)
X = np.arange(0, 50, 1)
ax.plot(X, datanormal[50:100], color=col1)
ax.plot(X, data3d[50:100], color=col2)
ax.plot(X, data3dshared[50:100], color=col3)
ax.plot(X, np.ones(50)*0.1, color=col4)
ax.plot(X, np.ones(50)*0.2, color=col5)
ax.set_title('Validation')
#ax.set_xlabel('Epoch')
#ax.set_ylabel('Accuracy')
ax.set_xlim(0, 50)
ax.set_ylim(0, 1)

#===============
# Third subplot
#===============

ax = fig.add_subplot(2, 3, 3)
X = np.arange(0, 50, 1)
ax.plot(X, datanormal[100:150], color=col1)
ax.plot(X, data3d[100:150], color=col2)
ax.plot(X, data3dshared[100:150], color=col3)
ax.plot(X, np.ones(50)*850, color=col4)
ax.plot(X, np.ones(50)*900, color=col5)
ax.set_title('Loss')
#ax.set_xlabel('Epoch')
ax.set_xlim(0, 50)
ax.legend(["Full","3D","3D - shared weights","3DATDC", "3DATDC - shared weights"])

#==============
#  second line
#==============

datanormal = pd.read_csv('07_05_Papernet4True_CIFAR100.csv', index_col = False, header = None)
data3d = pd.read_csv('07_05_Papernet4True3D_CIFAR100.csv', index_col = False, header = None)
data3dshared = pd.read_csv('07_05_Papernet4True3DSharedWeights_CIFAR100.csv', index_col = False, header = None)

#===============
#  First subplot
#===============
ax = fig.add_subplot(2, 3, 4)
X = np.arange(0, 50, 1)
ax.plot(X, datanormal[0:50], color=col1)
ax.plot(X, data3d[0:50], color=col2)
ax.plot(X, data3dshared[0:50], color=col3)
ax.plot(X, np.ones(50)*0.1, color=col4)
ax.plot(X, np.ones(50)*0.2, color=col5)
#ax.set_title('Training')
ax.set_xlabel('Epoch')
ax.set_ylabel('Papernet4\nAccuracy')
ax.set_xlim(0, 50)
ax.set_ylim(0, 1)

#===============
# Second subplot
#===============

ax = fig.add_subplot(2, 3, 5)
X = np.arange(0, 50, 1)
ax.plot(X, datanormal[50:100], color=col1)
ax.plot(X, data3d[50:100], color=col2)
ax.plot(X, data3dshared[50:100], color=col3)
ax.plot(X, np.ones(50)*0.1, color=col4)
ax.plot(X, np.ones(50)*0.2, color=col5)
#ax.set_title('Validation')
ax.set_xlabel('Epoch')
#ax.set_ylabel('Accuracy')
ax.set_xlim(0, 50)
ax.set_ylim(0, 1)

#===============
# Third subplot
#===============

ax = fig.add_subplot(2, 3, 6)
X = np.arange(0, 50, 1)
ax.plot(X, datanormal[100:150], color=col1)
ax.plot(X, data3d[100:150], color=col2)
ax.plot(X, data3dshared[100:150], color=col3)
ax.plot(X, np.ones(50)*800, color=col4)
ax.plot(X, np.ones(50)*900, color=col5)
#ax.set_title('Loss')
ax.set_xlabel('Epoch')
ax.set_xlim(0, 50)

### single line with parameters ###
'''
p1 = 
a = [1,2,5,6,9,11,15,17,18]

plt.hlines(1,1,20)  # Draw a horizontal line
plt.xlim(0,21)
plt.ylim(0.5,1.5)

y = np.ones(np.shape(a))   # Make all y values the same
plt.plot(a,y,'|',ms = 40)  # Plot a line at each location specified in a
plt.axis('off')
plt.show()
'''
plt.show()
```
