```python
%matplotlib qt

import pandas as pd

datanormal = pd.read_csv('3004_webNet_deeper.csv', index_col = False, header = None)
data3d = pd.read_csv('3004_webNet_deeper3d.csv', index_col = False, header = None)

import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

fig = plt.figure(figsize=plt.figaspect(0.4))


#===============
#  First subplot
#===============
ax = fig.add_subplot(1, 3, 1, projection='3d')
X = np.arange(0, 50, 1)
Y = [11, 20, 29, 38, 47, 56]
X, Y = np.meshgrid(X, Y)

Z = datanormal.iloc[0:6,:]
surf = ax.plot_surface(X, Y, Z, color='red',linewidth=1, antialiased=True, alpha = 0.6)

Z = data3d.iloc[0:6,:]
surf = ax.plot_surface(X, Y, Z, color='blue',linewidth=1, antialiased=True, alpha = 0.6)
ax.set_zlim(0, 1)
ax.zaxis.set_major_formatter('{x:.1f}')
ax.view_init(25, 20)


#===============
# Second subplot
#===============
ax = fig.add_subplot(1, 3, 2, projection='3d')

X = np.arange(0, 50, 1)
Y = [11, 20, 29, 38, 47, 56]
X, Y = np.meshgrid(X, Y)

Z = datanormal.iloc[10:16,:]
surf = ax.plot_surface(X, Y, Z, color='red',linewidth=1, antialiased=True, alpha = 0.6)

Z = data3d.iloc[10:16,:]
surf = ax.plot_surface(X, Y, Z, color='blue',linewidth=1, antialiased=True, alpha = 0.6)
ax.set_zlim(0, 1)
ax.zaxis.set_major_formatter('{x:.1f}')
ax.view_init(25, 20)


#===============
# Third subplot
#===============
ax = fig.add_subplot(1, 3, 3, projection='3d')

X = np.arange(0, 50, 1)
Y = [11, 20, 29, 38, 47, 56]
X, Y = np.meshgrid(X, Y)

Z = datanormal.iloc[20:26,:]
surf = ax.plot_surface(X, Y, Z, color='red',linewidth=1, antialiased=True, alpha = 0.6)

Z = data3d.iloc[20:26,:]
surf = ax.plot_surface(X, Y, Z, color='blue',linewidth=1, antialiased=True, alpha = 0.6)
ax.set_zlim(600, 2500)
#ax.zaxis.set_major_formatter('{x:100.0f}')
ax.view_init(25, 20)

plt.show()
```
