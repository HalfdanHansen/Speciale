```python
%matplotlib qt

import pandas as pd

datanormal = pd.read_csv('3004_webNet_deeper.csv', index_col = False, header = None)
data3d = pd.read_csv('3004_webNet_deeper3d.csv', index_col = False, header = None)

import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib.patches as mpatches
from numpy import random
from matplotlib.font_manager import FontProperties

fig = plt.figure(figsize=plt.figaspect(0.4))

fig.suptitle('WebNet of Different Depth')

col1, col2 = cm.jet(np.array([0.9, 0.1]))
fontP = FontProperties()
fontP.set_size('small')

#===============
#  First subplot
#===============
ax = fig.add_subplot(1, 3, 1, projection='3d')
X = np.arange(0, 50, 1)
Y = [11, 20, 29, 38, 47, 56]
X, Y = np.meshgrid(X, Y)

Z = datanormal.iloc[0:6,:]
surf = ax.plot_surface(X, Y, Z, color=col1,linewidth=1, antialiased=True, alpha = 0.6)

Z = data3d.iloc[0:6,:]
surf = ax.plot_surface(X, Y, Z, color=col2,linewidth=1, antialiased=True, alpha = 0.6)
ax.set_zlim(0, 1)
ax.zaxis.set_major_formatter('{x:.1f}')
ax.view_init(25, 20)
ax.set_title('Training')
ax.set_xlabel('Epoch')
ax.set_ylabel('Depth')
ax.set_zlabel('Accuracy')

#===============
# Second subplot
#===============
ax = fig.add_subplot(1, 3, 2, projection='3d')

X = np.arange(0, 50, 1)
Y = [11, 20, 29, 38, 47, 56]
X, Y = np.meshgrid(X, Y)

Z = datanormal.iloc[10:16,:]
surf = ax.plot_surface(X, Y, Z, color=col1,linewidth=1, antialiased=True, alpha = 0.6)

Z = data3d.iloc[10:16,:]
surf = ax.plot_surface(X, Y, Z, color=col2,linewidth=1, antialiased=True, alpha = 0.6)
ax.set_zlim(0, 1)
ax.zaxis.set_major_formatter('{x:.1f}')
ax.view_init(25, 20)
ax.set_title('Valdidation')
ax.set_xlabel('Epoch')
ax.set_ylabel('Depth')
ax.set_zlabel('Accuracy')

col1_patch = mpatches.Patch(color=col1, label='Full model')
col2_patch = mpatches.Patch(color=col2, label='3D CD Decomposed Model')
plt.legend(handles=[col1_patch, col2_patch], bbox_to_anchor=(0.5, -0.1), loc = 9, prop=fontP)

#===============
# Third subplot
#===============
ax = fig.add_subplot(1, 3, 3, projection='3d')

X = np.arange(0, 50, 1)
Y = [11, 20, 29, 38, 47, 56]
X, Y = np.meshgrid(X, Y)

Z = datanormal.iloc[20:26,:]
surf = ax.plot_surface(X, Y, Z, color=col1,linewidth=1, antialiased=True, alpha = 0.6)

Z = data3d.iloc[20:26,:]
surf = ax.plot_surface(X, Y, Z, color=col2,linewidth=1, antialiased=True, alpha = 0.6)
ax.set_zlim(600, 2500)
#ax.zaxis.set_major_formatter('{x:100.0f}')
ax.view_init(25, 20)
ax.set_title('Loss')
ax.set_xlabel('Epoch')
ax.set_ylabel('Depth')

plt.show()
```
