```python
import pandas as pd
```

```python
initDF = pd.read_csv('0704_CIFAR10_no_pass_net_D4DD_rank123.csv', header = None) 
```

```python
sep = (12*5)
```

```python
trainDF = initDF[0:sep]
testDF = initDF[sep+1:sep*2]
lossDF = initDF[sep*2:-1]
```

```python
trainDF
```

```python
def f(x):
    if x.last_valid_index() is None:
        return np.nan
    else:
        return x[x.last_valid_index()]
```

```python
trainDF['last'] = trainDF.apply(f, axis=1)
testDF['last'] = testDF.apply(f, axis=1)
lossDF['last'] = lossDF.apply(f, axis=1)
```

```python
c1 = 0
c2 = 5  
train_list = [] 
test_list = [] 
loss_list = [] 
while c2 <= 60:
    temp_train = trainDF["last"].values[c1:c2]
    temp_test = testDF["last"].values[c1:c2]
    temp_loss = lossDF["last"].values[c1:c2]
    train_list.append(temp_train)
    test_list.append(temp_test)
    loss_list.append(temp_loss)
    c1 = c2
    c2 = c2+5
train_box = pd.DataFrame(train_list).transpose()
test_box = pd.DataFrame(test_list).transpose()
loss_box = pd.DataFrame(loss_list).transpose()
```

```python
import numpy as np
```

```python
train_boxplot = train_box.boxplot(return_type='axes')
```

```python
test_boxplot = test_box.boxplot()
```

```python
loss_boxplot = loss_box.boxplot()
```
